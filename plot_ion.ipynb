```python
data['year'] = [d.year for d in data.Data]
years = data['year'].unique()
minimos=[]
for ano in years:
    a=(data['Vazao']==data.loc[data['year']==ano]['Vazao'].min()) & (data['year']==ano)
    testegeral=data[a].set_index('Data')
    minimos.append(testegeral['Vazao'])
mintotal=pd.concat(minimos)
```

```python
mintotal.diff()
```

```python
fig,ax = plt.subplots()
ax.plot(data['Data'],data['ION'], color="red",label='ION')
ax.set_xlabel("Ano",fontsize=14)
ax.set_ylabel("ION",color="red",fontsize=14)
ax2=ax.twinx()
ax2.plot(mintotal.diff(), color = 'black',label='VazÃ£o')
ax2.set_ylabel("VariaÃ§Ã£o de VazÃ£o (mm)",color="black",fontsize=14)
yabs_max1 = abs(max(ax.get_ylim(), key=abs))
ax.set_ylim(ymin=-yabs_max1, ymax=yabs_max1)
ax2.set_ylim(ymin=-100, ymax=100)
plt.rcParams['figure.figsize'] = [15, 7]
plt.title('ENSO X VariaÃ§Ã£o de vazÃµes mÃ­nimas - PindarÃ©-Mirim')
plt.grid(linestyle='--',linewidth=1.2)
lines_1, labels_1 = ax.get_legend_handles_labels()
lines_2, labels_2 = ax2.get_legend_handles_labels()
lines = lines_1 + lines_2
labels = labels_1 + labels_2
ax.legend(lines, labels, loc=0,shadow=True)
plt.show()
```
